# Example 01

$$\max Z = 3x_1 + 4x_2 $$

\begin{align}
x_1 + x_2 & \leq 450\\
2x_1 + x_2 & \leq 600\\[4mm]
x_1, x_2 & \geq 0
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 500, 50)
```

```python
eq1 = 450 - x    # x2 = 450 - x1
eq2 = 600 - 2*x  # x2 = 600 - 2x1
```

```python
zeq = lambda x1, z:(z - 3*x)/4   # x2 = (Z - 3x1)/4
```

## Plot

```python
plt.figure(figsize=(8,8))

plt.axvline(0, color="0.4")
plt.axhline(0, color="0.4")
# x, f(x)
plt.plot(x, eq1, lw=2, label=r"$x_1 + x_2 \leq 450$")
plt.plot(x, eq2, lw=2, label=r"$2x_1 + x_2 \leq 600$")

plt.plot(x, zeq(x, 0), color="magenta", ls="--", lw=2, alpha=0.4, label=r"$\max\, Z = 3x_1 + 4x_2$")
for i in np.linspace(500, 1500, 4):
    plt.plot(x, zeq(x, i), color="magenta", lw=2, ls="--", alpha=0.4)

plt.legend(fontsize=15)
plt.xlim(-10, 500)
plt.ylim(-10, 500)
plt.show()
```

```python
points = [(0,0), (0, 450), (150, 300), (300, 0)]  # intersecting points
zvalues = [3*x1 + 4*x2 for x1,x2 in points]
```

```python
print(zvalues)
```

```python
best_index, best_value = max(enumerate(zvalues), key=lambda t:t[1])
```

```python
print(best_index)
print(best_value)
```

```python
best_point = points[best_index]
print(best_point)
```

```python
coordinates = list(zip(*points))
print(coordinates)
```

```python
# plot the feasible region
plt.figure(figsize=(8,8))
plt.axvline(0, color="0.4")
plt.axhline(0, color="0.4")

plt.plot(x, eq1, lw=2, label=r"$x_1 + x_2 \leq 450$")
plt.plot(x, eq2, lw=2, label=r"$2x_1 + x_2 \leq 600$")

plt.plot(x, zeq(x, 0), color="magenta", ls="--", lw=2, alpha=0.4, label=r"$\max\, Z = 3x_1 + 4x_2$")
for i in np.linspace(500, 1500, 4):
    plt.plot(x, zeq(x, i), color="magenta", lw=2, ls="--", alpha=0.4)

for point in points:
    plt.plot(*point, marker="o", color="k", ms=7)

plt.fill(*coordinates, facecolor="yellow", edgecolor="0.6", alpha=0.4 )

plt.plot(*best_point, marker="*", ms=15, color="red")
plt.plot(x, zeq(x, best_value), color="magenta", ls="--", lw=2, alpha=0.6)

plt.legend(fontsize=15, loc="upper right")
plt.xlim(-10, 500)
plt.ylim(-10, 500)
plt.show()
```
